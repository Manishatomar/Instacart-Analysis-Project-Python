## Step 3 You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column. In a markdown cell beneath your histogram, describe what the histogram demonstrates.

```python
#creating histogram for 'order_hour_of_day'

hist = instacart['order_hour_of_day'].plot.hist(bins = 24)
```


The histogram shows peaks at specific hours, indicating when most orders are placed. Peaks at midday or early evening suggest customers prefer shopping during lunch or after work. The distribution shape reveals if orders are concentrated at certain times or spread evenly. A normal distribution points to consistent shopping around a particular time, while a skewed distribution indicates preferences for specific times, like afternoons or evenings.
