# 02. Creating a Bar Chart

Creating a bar chart is similar to printing column frequencies, but visual. Bar charts show how observations are distributed across categories. Let’s create one for the “orders_day_of_week” column.

```python
#Creating bar chart

df['orders_day_of_week'].value_counts().plot.bar()
```

Here’s a breakdown of the code: Use basic indexing to select the column and then call value_counts() to get frequencies. Add plot.bar() to create a bar chart from these frequencies. Always use this function on the column’s frequency, not the column itself, to avoid long processing times. The chart orders observations from largest to smallest, but you can change this with the sort_index() function if needed.

```python
#indexing by order of days of week. The sort_index() function will sort the bars in your chart according to their index order (rather than 
# largest to smallest).

bar = df['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# can reorder columns to appear in their indexed order when printing frequencies too. While the numbers stay the same, the order in which they
# appear has changed. This can be useful if data would be easier to understand in its original order. Days of the week/months of the year, 
# for instance, would be better understood if they were listed chronologically

df['orders_day_of_week'].value_counts()
```

```python
df['orders_day_of_week'].value_counts().sort_index()
```

# 04. Creating Scatterplots

Scatterplots reveal relationships between two variables and help identify outliers. Let’s plot the “prices” column against itself to check for outliers.

```python
# Creating Histogram

sns.scatterplot(x = 'prices', y = 'prices',data = df)
```

Create scatterplots using the scatterplot() function from seaborn (sns). Plot the “prices” column on both axes. Data points should cluster around the mean of 12, but you’ll see outliers near 20,000 and 100,000. The 100,000 value matches the maximum from your checks. To explore further, check for values above a typical supermarket price, such as $100, which might indicate outliers. Use the following code to perform this check:

```python
#finding prices over $100

df.loc[df['prices'] > 100]
```

The output shows 5,127 rows with outliers (prices over $100). The odd value of $14,900.00 and the $99,999.00 are skewing the charts.

```python
# For now, let’s mark them as missing as they don’t make sense in terms of other values in column.To turn them into NaNs,use following code:
#assigning null values to outliers

df.loc[df['prices'] >100, 'prices'] = np.nan
```

The command assigns np.nan (from NumPy) to values above 100 in the “prices” column, marking them as missing.

```python
# Once finished, check whether the replacement was successful by running a max-value check on your “prices” column 
#finding max value in prices

df['prices'].max()
```

With outliers replaced by missing values, the maximum price is now a realistic $25. A $25 item is far more likely than $14,900 or $99,999. Let’s create the histogram of the “prices” column again.

```python
#creating histogram

hist = df['prices'].plot.hist(bins = 25)
```

The results are clearer: most products are priced between $1 and $15, with a few in the $15 to $25 range. To add detail, increase the number of bins to 70 for a histogram with more, thinner columns.

# 06. Sampling Data

To create a representative subset of your data, you need more than just the first 2,000,000 rows, and they must be randomized. Random sampling from the entire dataframe is necessary.Start by generating random samples. Split your dataframe into two samples with similar distributions, such as 80/20 or 70/30. In a 70/30 split, the smaller sample should statistically resemble the larger one but have fewer rows, accurately representing the whole dataframe. Generate random numbers between 0 and 1 for each row. For a 70/30 split, rows with numbers less than 0.7 go into one sample, and those with numbers greater than 0.7 go into the other, effectively splitting the dataframe.

```python
#setting the seed

np.random.seed(4)
```

```python
dev
```

In first line, you set a "seed" with random.seed(4) to ensure Python generates same random numbers every time, making example reproducible. In the Third line, you create a list called dev to store boolean values, outputting True or False based on whether each value is less than or equal to 0.7.

You're using np.random.rand(), which generates a list of random floating numbers between 0 and 1, following a uniform distribution. The np stands for NumPy, random creates random numbers, and rand ensures a uniform distribution. By passing len(ords_prods_merge) into the function, you generate random numbers for all 32,404,859 rows in your dataframe. Here's a simple example of np.random.rand() to illustrate this.

```python
# The function generates a list of random numbers between 0 and 1. If the argument is 10, it produces 10 random numbers. 
# In your code, the argument is the length of your dataframe, so it generates 32,434,212 random numbers
np.random.rand(10)
```

To recap the split code:

Set the seed for the random number generator.
Create 32,404,859 random numbers (matching the length of your dataframe) between 0 and 1.
Compare each number to 0.7 (less than or greater than).
Store the boolean results (True/False) in a new list, dev.
You now have boolean list and dataframe with 32,434,212 rows to split into two samples. Here's how:

```python
#split into 'big' and 'small'

big = df[dev]
small = df[~dev]
```

You’ll create two dataframes, big and small, from ords_prods_merge. Store 70% of the rows in big and 30% in small based on your dev list. Rows with True values go to big, and rows with False values go to small. The tilde in the second line negates the argument, capturing the False values. Here's the workflow in Jupyter:

```python
# Create a list holding True/False values to the test np.random.rant() <= 0.7
# assigning 'dev'

dev = np.random.rand(len(df)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big

big = df[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = df[~dev]
```

Before creating your chart, check your results. If big is 70% and small is 30% of ords_prods_merge, their combined rows should equal the original dataframe. Use the len() function to verify this.

```python
#checking results
len(df)
```

```python
len(big) + len(small)
```
